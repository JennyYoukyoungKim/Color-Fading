```python
import pandas as pd

df = pd.read_csv(r"C:\Users\jenny.kim-es\Downloads\CI_Logarithmic.csv")

print(df)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
import pylab
#sns.regplot(data=df, x="Reflectance", y="Mean", logx=True, ci=90)

import numpy as np
from scipy.stats import pearsonr

r1, pvalue = pearsonr(np.log(df['Reflectance']), df['Mean'])
r2 = r1 *r1
g=sns.regplot(data=df, x='Reflectance', y='Mean', logx=True, ci=90, label=f'R^2 = {r2:.5f}')
g.legend()

ax = g
lower = [line.get_ydata().min() for line in ax.lines]
upper = [line.get_ydata().max() for line in ax.lines]

print (lower)
print (upper)
plt.savefig('logarithmic_ci.jpg')
```
