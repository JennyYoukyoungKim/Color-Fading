```r
df=read.csv('C:\\Users\\jenny.kim-es\\Downloads\\CI_logarithmic.csv')
```

```r
head(df)
```

```r
df<-df[order(df$Reflectance),]
```

```r
loess_res<-loess(Mean~Reflectance,data=df,span=0.3,degree=2)

# Generate predictions from the LOESS model
predictions <- predict(loess_res, newdata = df)
predictions <- cbind(df, predictions)
predictions <- predictions[order(predictions$Reflectance), ]

# Create a scatter plot of the original data
plot(
  predictions$Reflectance,
  predictions$Mean, 
  col = "blue", 
  main = "LOESS Regression: Unveiling the Magic with mtcars", 
  xlab = "Weight (1000 lbs)", 
  ylab = "Miles Per Gallon"
)

# Add the LOESS curve to the plot
lines(predictions$predictions, col = "red", lwd = 2)
```
